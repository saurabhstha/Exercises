```python
#import dependecies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#load datasets
df = pd.read_csv('birth-rates-yearly.csv')

df_brate = pd.read_csv('birth-rate.csv')

df_state = pd.read_csv("crimerates-by-state-2005.csv")
```

```python
#delete redundant data
df = df.iloc[0:49,:]
```

```python
#Scatterplot for yearly birth rate
plt.rcParams['figure.figsize'] = [8, 5]
sns.scatterplot(x = 'year', y ='rate', data=df)
plt.xlabel("Year")
plt.ylabel('Birth Rate')
plt.title('Yearly Birth Rate')
```

```python
#plotting for density plot
df.rate.plot.density(color='green')
plt.title("Density Plot For Birth Rate")
```
